```python
# Subset the data based on the criteria that ST_FIPS equals 42
recent_scorecard_data[recent_scorecard_data['ST_FIPS'] == 42]
```

* Huh? Why does this work? What is going on here?

### What is the distribution of public, private, and for-profit schools in Pennsylvania?

* Go to the [data dictionary](https://docs.google.com/spreadsheets/d/1gSnivIY8KlMlqID0uWPyVhdPnLEgxpTpZwsMzBwlo28/edit?usp=sharing)
* What variable do you think we need to answer this question? *Hint: It is close to the top of the dictionary.*

* The data dictionary tells us the `CONTROL` column has the information we need.
* How do we *count* the various *values* in this column? Remember what we did a few weeks ago?

```python
# Use the value_counts function to see what values appear
pa_schools['CONTROL'].value_counts()
```

* Hrm, that isn't very pretty. Let's map those values to *humane* names
* We can use the `replace()` function to change the numbers to names

```python
# Create a python dictionary mapping the number to a more readable name
# We get this mapping from the data dictionary
control_values = {1:"Public",
                  2:"Private nonprofit",
                  3:"Private for-profit"}

# Use the replace function and then select just the `CONTROL` column
pa_school_types = pa_schools.replace({"CONTROL":control_values})['CONTROL']
pa_school_types.value_counts()
```

* Cool, but this is a lecture about data visualization so let's make an annotated bar chart.

```python
# Generate a bar chart with a label and the x-labels horizontal (the default is vertical)
ax = pa_school_types.value_counts().plot(kind="bar",
                                         title="Schools types in Pa",
                                         rot=0)

```

* This is great, but why don't we label each category

```python
# Generate a bar chart with a label and the x-labels horizontal (the default is vertical)
ax = pa_school_types.value_counts().plot(kind="bar",
                                         title="Schools types in PA",
                                         rot=0)

# Annotate each "patch" with  
for bar in ax.patches:
    
    ax.text(bar.get_x() + (bar.get_width()/3), 
            bar.get_height() - 10, 
            int(bar.get_height()),
            weight="bold",
            color="white")
    
```

```python
# Generate a bar chart with a label and the x-labels horizontal (the default is vertical)
ax = pa_school_types.value_counts().plot(kind="bar",
                                         title="Schools types in Pa",
                                         rot=0)

# Loop over each bar ("patch")
for bar in ax.patches:
    # us the ax.text function to add text annotations to the chart
    ax.text(bar.get_x() + (bar.get_width()/3), # Set the x position
            bar.get_height() - 10,             # Set the y position
            int(bar.get_height()),             # Set the value
            weight="bold")                     # Make it bold for readability
    
# Add axes labels
ax.set_xlabel("Sector")
ax.set_ylabel("Count")
```

* We did it! Not very intuitive, right? That's matplotlib for you. Sigh.
* Sometimes the best way to figure out how to do something is google around and look at other people's code.

### Cost for students from various socio-economic backgrounds at Pitt & CMU

```python
# create nice labels
income_levels = {"NPT41_PUB":"0-30k",
                 "NPT42_PUB":"30k-48k",
                 "NPT43_PUB":"48k-75k",
                 "NPT44_PUB":"75k-110k",
                 "NPT45_PUB":"110k+",
                 "NPT41_PRIV":"0-30k",
                 "NPT42_PRIV":"30k-48k",
                 "NPT43_PRIV":"48k-75k",
                 "NPT44_PRIV":"75k-110k",
                 "NPT45_PRIV":"110k+"}
```

```python
# Note, we have different columns for public vs. private
net_price_per_level_pub = """NPT41_PUB
NPT42_PUB
NPT43_PUB
NPT44_PUB
NPT45_PUB""".split("\n")
net_price_per_level_priv = """NPT41_PRIV
NPT42_PRIV
NPT43_PRIV
NPT44_PRIV
NPT45_PRIV""".split("\n")
```

```python
# Combine them now that they have the same index
pitt_and_cmu_income_levels = pd.concat([pitt_income_levels, cmu_income_levels], axis=1)
pitt_and_cmu_income_levels
```

```python
# PLOTIT!
ax = pitt_and_cmu_income_levels.plot(kind="bar",
                                     title="Average cost per income level",
                                     rot=0)
ax.set_xlabel("Income Level")
ax.set_ylabel("Cost")
```
